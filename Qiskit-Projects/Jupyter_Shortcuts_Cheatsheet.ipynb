## 🧠 Jupyter Notebook Shortcut Cheatsheet

### 🎯 Navigation & Cell Execution

| Action                          | Shortcut                |
|---------------------------------|-------------------------|
| ▶️ Run cell & go to next        | `Shift + Enter`         |
| ⏬ Run cell & stay              | `Ctrl + Enter`          |
| ➕ Insert cell below            | `B`                     |
| ➕ Insert cell above            | `A`                     |
| ❌ Delete current cell          | `D` (twice)             |
| 🔁 Undo delete                  | `Z`                     |
| ✏️ Edit mode                    | `Enter`                 |
| 🟦 Command mode                 | `Esc`                   |
| 🔼 Go to cell above             | `↑`                     |
| 🔽 Go to cell below             | `↓`                     |

---

### 🪄 Output Control

- Click ▾ or ▸ icon near output to collapse/expand output manually
- Programmatically clear output:
  ```python
  from IPython.display import clear_output
  clear_output(wait=True)
  ```

---

### ✍️ Cell Type & Formatting

| Action                     | Shortcut    |
|----------------------------|-------------|
| 🧮 Change to Code          | `Y`         |
| 📄 Change to Markdown      | `M`         |
| 🧾 Toggle line numbers     | `L`         |
| 🧮 Math (Markdown):        | `$x^2 + y^2$` |

---

### 🔁 Copy, Paste, Move Cells

| Action                   | Shortcut      |
|--------------------------|---------------|
| 📋 Copy cell             | `C`           |
| ✂️ Cut cell              | `X`           |
| 📥 Paste cell below     | `V`           |
| ⬆️ Move cell up         | `Shift + K`   |
| ⬇️ Move cell down       | `Shift + J`   |

---

### 📘 Tip:
- **Green border** = Edit mode
- **Blue border** = Command mode

Happy computing! ⚛️🔬
